# Planet:Star Ratio Frequency

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
hist, xedges, yedges = np.histogram2d(Star, Planet, bins = 4, range = [[0,4], [0,4]])
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort = 'average')

plt.show()
```
